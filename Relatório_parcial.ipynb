# **Computação Quântica: Teoria e Aplicações**

## O que é?  
A computação quântica é uma área em que se aplicam os princípios da mecânica quântica. Nela, os supercomputadores podem resolver problemas complexos que um computador comum levaria milhões de anos para resolver, em apenas alguns segundos. Falar em computação clássica remete imediatamente aos bits, que é um sistema binário com dois estados $|0\rangle, |1\rangle\$, mas na computação quântica usa-se um sistema quântico de dois níveis chamado de bit quântico ou simplesmente qubit.

## O que é Qubit?  
É a menor unidade de informação em um computador quântico, em que há dois graus de liberdade, nos quais uma superposição dos estados da base computacional clássica pode ocorrer.

## Representação

* Base Computacional:
$$
|0\rangle = \begin{bmatrix} 1 \\ 0 \end{bmatrix}, \quad |1\rangle = \begin{bmatrix} 0 \\ 1 \end{bmatrix}.
$$

## O que seria essa superposição?  
Como sabemos, quando se fala em estado quântico, podemos escolher uma base para representar os estados e temos que levar em conta combinações lineares dos estados da base, chamadas superposições.

* Superposição da Base Computacional:
$$
|\psi\rangle = \alpha |0\rangle + \beta |1\rangle \quad \text{onde} \quad \alpha, \beta \in \mathbb{C} \ (\text{números complexos}). \tag{1}
$$

Aqui, $\alpha\ e \beta\$ são números complexos que cumprem a *condição de normalização*:

$$|\alpha|^2 + |\beta|^2 = 1. $$

Essa condição garante que a soma das probabilidades de medir os dois estados possíveis seja 1.

Um princípio fundamental da Mecânica Quântica diz respeito aos resultados das medidas, o que chamamos *regra de Born*: quando você mede um qubit no estado $|\psi\rangle$ (*Eq.1*), existe uma probabilidade $|\alpha|^2$ de que o resultado da medição seja $|0\rangle$ e uma probabilidade $|\beta|^2$ de que o resultado seja $|1\rangle$. Após essa medição, o qubit assume o estado $|0\rangle$ ou o estado $|1\rangle$.


$$
P_\psi(0) = |\alpha|^2 ,\
P_\psi(1) = |\beta|^2.
$$

# **Portas Quânticas**


Assim como na computação clássica utilizamos portas lógicas (como AND, OR, NOT), na computação quântica utilizamos portas quânticas para manipular os estados de qubits. Essas portas são representadas por operações unitárias, os quais são operadores lineares que cumprem as seguintes condições:

$$
U^\dagger U = U U^\dagger =I.
$$

Destacamos que os operadores unitários (logo, as portas quânticas) preservam a norma dos estados quânticos. Em termos práticos, uma porta quântica realiza uma transformação que "move" o estado do qubit para um novo ponto na Esfera de Bloch apresentado anteriormente.

* Da mesma forma um qubit no estado $|1⟩$ a saída será $|0⟩$.

$$
X = NOT= \begin{bmatrix}
0 & 1\\
1 & 0

### Porta ($Z$):

A porta $Z$ é uma porta unitária que atua somente em um qubit, agindo na base computacional da seguinte forma:

* Mapeia um qubit no estado $|1\rangle$ para $-|1\rangle$.

* Deixa o qubit no estado $|0\rangle$ inalterado.

A operação é descrita pela matriz $Z$:

$$
Z = \begin{bmatrix}
1 & 0\\
0 & -1
\end{bmatrix}
$$

* Na Esfera de Bloch, essa porta corresponde a uma rotação de 180° ao redor do eixo $Z$.

### Porta ($Y$)

A porta $Y$ realiza uma operação unitária específica em um estado quântico, agindo na base computacional da seguinte forma:

* Um qubit no estado $|1\rangle$ sera mapeado para o estado $-i|0\rangle$.

* Um qubit no estado $|0\rangle$ sera mapeado para o estado $i|1\rangle$.

A operação é descrita pela matriz $Y$:

$$
Y = \begin{bmatrix}
0 & -i\\
i & 0
\end{bmatrix}
$$

* Na Esfera de Bloch, a porta $Y$ corresponde a uma rotação de 180° ao redor do eixo $Y$.

### Porta de Fase ($S$):

A porta de fase aplica uma rotação no estado quântico, ajustando a fase do estado $|1⟩$ enquanto matem o estado $|0⟩$ inalterado.

* Um qubit no estado $|0\rangle$ permanece $|0\rangle$
* Um qubit no estado $|1\rangle$ é mapeado para $i|1\rangle$

A operação é descrita pela matriz $S$:

$$
S = \begin{bmatrix}
1 & 0\\
0 & i
\end{bmatrix}
$$

Na Esfera de Bloch, a porta $S$ realiza uma rotação de 90° ao redor do eixo Z.

### Porta Hadamard ($H$):

A porta Hadamard transforma um estado quântico da base computacional em uma superposição de estados.

* Um qubit no estado $|0\rangle$ será mapeado para

$$
\frac{|0\rangle + |1\rangle}{\sqrt{2}}
$$

* Um qubit no estado $|1\rangle$ será mapeado para

$$
\frac{|0\rangle - |1\rangle}{\sqrt{2}}
$$

A operação é descrita pela matriz $H$:

$$
H = \frac{1}{\sqrt{2}}\begin{bmatrix}
1 & 1\\
1 & -1
\end{bmatrix}
$$

Na Esfera de Bloch, a porta Hadamard corresponde a uma combinação de rotações, criando estados de superposição.

### Algoritmo de Deutsch

$\quad$ O Algoritmo de Deutsch é um dos primeiros exemplos de como a computação quântica pode resolver certos problemas de maneira mais eficiente do que a computação clássica.


#### O problema:

$\quad$ Dada uma função binária $f: \left\lbrace 0,1 \right\rbrace \rightarrow \left\lbrace 0,1 \right\rbrace$ determina se $f(x)$ é constante (retorna o mesmo valor para todas as entradas) ou balanceada (retorna 0 para uma entrada e 1 para outra)

* Na computação clássica, seriam necessárias duas chamadas à função $f$, uma para $f(0)$ e outra para $f(1)$.

* Na computação quântica, isso pode ser resolvido com uma única chamada à função $f$, mediante o uso da superposição e interferência dos estados quânticos.

#### Passo a Passo do circuito de Deutsch


* Estado inicial: Preparamos dois qubits no estado $|0\rangle$ e $|1\rangle$

$$
|\psi_0\rangle = |0\rangle \otimes |1\rangle
$$

* Porta Hadamard: Aplicamos a porta Hadamard nos dois qubits para criar uma superposição


$$
H |0⟩ = \frac{|0\rangle + |1\rangle}{\sqrt{2}}
\\
H |1⟩ = \frac{|0\rangle - |1\rangle}{\sqrt{2}}
$$

Após essa etapa, o estado do sistema é:

$$
|\psi_1\rangle = \frac{|0\rangle + |1\rangle}{\sqrt{2}} \otimes \frac{|0\rangle - |1\rangle}{\sqrt{2}}
$$

* Aplicação do operador $U_f$, um oráculo que implementa $f$ no circuito quântico: Para construir e executar o algoritmo, precisamos de uma implementação quântica de f, que em geral tem a seguinte forma

$$
U_f : | x \rangle | y \rangle = | x \rangle | y \oplus f(x) \rangle
$$


* Aplicação da Porta Hadamard no Primeiro Qubit: Após processar $U_f$, aplicamos novamente a porta Hadamard no primeiro qubit para "recombinar" as amplitudes de probabilidade.


* Medição: O resulatado da medição no qubit um indica de $f(x)$ é constante $|0\rangle$ ou balanceada $|1\rangle$.

### Emaranhamento

$\quad$ Emaranhamento é uma característica de alguns estados de múltiplos objetos (como um par de elétrons ou fótons) que compartilham um único estado quântico. Assim como fios em um emaranhado de lã, objetos entrelaçados não podem ser descritos como entidades independentes. Por exemplo, suponha que dois elétrons, A e B, estejam entrelaçados. Se o elétron A estiver na Terra e o elétron B estiver em Andrômeda, o estado quântico de A estará diretamente relacionado ao estado quântico de B. Se algo alterar ou medir o estado de A, o estado de B será instantaneamente afetado, mesmo estando a anos-luz de distância.

* Swap de Emaranhamento

$\quad$ O Protocolo de Emaranhamento Swap mostra que é possível emaranhar dois sistemas quânticos, digamos A e B, que nunca interagiram diretamente. Isso pode acontece no protocolo porque os sistemas A e B estão inicialmente emaranhados com outros sistemas, digamos C1 e C2, os quais acabam sendo medidos (ou seja, interagindo). Portanto, o protocolo tem o efeito de trocar o emaranhamento de pares de sistemas quânticos mediante a interação de suas partes.

### Transformada Quântica de Fourier

$\quad$ A transformada quântica de Fourier (QFT) é uma operação fundamental na computação quântica que desempenha um papel importante em muitos algoritmos quânticos, incluindo o algoritmo de Shor para fatorar números grandes e estimativa de fase quântica.



*	A TFQ consiste em uma implementação da *Transformada de Fourier Discreta*:
$$
f: \mathbb{C}^{\times N} \rightarrow \mathbb{C}^{\times N},\ \mathbf{y} = f(\mathbf{x}) \leadsto y_j = \sum_{k=0}^{N-1} e^{-2\pi i \frac{kj}{N}}x_k.
$$
*	A TFQ é muito similar à transformada de Hadamard, com a diferença essencial sendo a introdução de fases:
$$
\begin{split}
H_n: \mathcal{H}_2^{\otimes n} \rightarrow \mathcal{H}_2^{\otimes n},\ H_n |x_{n-1\dots x_1x_0} \rangle &= (H|x_{n-1}\rangle) \otimes \dots \otimes (H|x_1\rangle) \otimes (H|x_0\rangle)\\
&=(|0 \rangle + (-1)^{x_n})|1\rangle \otimes \dots \otimes (|0 \rangle + (-1)^{x_1})|1\rangle \otimes (|0 \rangle + (-1)^{x_0})|1\rangle)\\
&=...
\end{split}
$$


**Definição de TFQ**: Para $n \in \mathbb{N},\ N = 2^n$

$$
F_n: \mathcal{H}_2^{\otimes n} \rightarrow \mathcal{H}_2^{\otimes n},\ F_n|x\rangle:= |y\rangle,
$$
onde:
$$
|x\rangle = \sum_{k=0}^{2^n-1} x_k | k \rangle \leadsto |y \rangle = \sum_{j=0}^{2^n-1} y_j | j \rangle
$$
e
$$
y_j = \frac{1}{\sqrt{2^n}}\sum_{k=0}^{2^n-1} e^{2\pi i \frac{kj}{2^n}}x_k.
$$
Destacam-se as diferenças com a *transformada de fourier discreta*:
*   sinal positivo nos expoentes e
*   fator de normalização.

<font color="gray">**TFQ na base computacional**: a TFQ de um estado da base computacional resulta em uma superposição uniforme de todos os estados da base computacional:
$$
F_n| k '
$$
A medida de $F_n| k \rangle $ na base computacional tem probabilidade igual a $1/2^n$ de resultar em qualquer um dos estados $|j\rangle$:
$$
\langle j | F_n | k\rangle = \left|  \frac{e^{2\pi i \frac{jk}{2^n}}}{\sqrt{2^n}} \right|^2 = \frac{1}{2^n},\ \ \ \forall k, j \in \left\lbrace 0,1, \dots, 2^n-1 \right\rbrace.
$$
></font>

[F_n] = \frac{1}{\sqrt{2^n}}\begin{pmatrix}

1 & \omega & \omega^2 & \omega^3 & \dots & \omega^{2^n-1}\\

$$

# **QAOA**

Definimos um espaço de qubits associado ao problema e um hamiltoniano associado à função custo:
$$
\mathcal{H} = \mathcal{H}_1^{\otimes n_V} = \mathrm{spam}\left\{ |z_{n_V - 1}\dots z_1z_0 \rangle;\  z_0, z_1, \dots,  z_{n_V - 1} \in \left\{ 0,1\right\}\right\}
$$
e
$$
H_C = \frac{1}{2}\sum_{i<j =0}^{n-1} w_{ij} (I - Z_iZ_j),
$$
onde $Z_k$s denota o operador de $\mathcal{H}$ que atua no $k$-ésimo qubit como a porta $Z$:
$$
Z_k |z_{n_V - 1}\dots z_1z_0 \rangle = (-1)^{j_k} |z_{n_V - 1}\dots z_1z_0  \rangle.
$$


*Observamos que $H_C$ é um operador diagonalizado na base computacional, com autovalores correspondendo às imagens da função custo:*
$$
H_C | z_{n_V - 1} \dots z_1 z_0 \rangle = C(\mathbf{z}) | z_{n_V - 1} \dots z_1 z_0 \rangle,\ \forall \mathbf{z} = (z_0, z_1, \dots z_{n_V - 1}) \in \left\{0,1 \right\}^{\times n_V}.
$$

◼

Agora, definimos o seguinte operador parametrizado associado ao hamiltoniano do custo:

$$
U(H_C,\gamma) := e^{-i\gamma H_C} = \prod_{0 \le i<j \le n-1} e^{-i\gamma w_{ij}(I - Z_i Z_j)/2}
$$

Também definimos o operador mixer parametrizado:
$$
U(B,\beta) = e^{-i\beta B} = \prod_{j=0}^{n-1} e^{-i\beta X_j},\ \text{onde}\ B = \sum_{j=0}^{n-1} X_j.
$$

>>> **Teorema (QAOA)** Considere:
$$
\
$$
i) o estado emaranhado
$$
| s \rangle = H^{\otimes} | 0 \rangle^{\otimes n} = \frac{1}{\sqrt{2^n}} \sum_{k=0}^{2^{n_V}-1} | k \rangle;
$$
ii) a evolução do estado $| s \rangle $por uma sequência finita de ações dos operadores de custo e mixer (com diferentes parâmetros)
$$
| \mathbf{\gamma}, \mathbf{\beta} \rangle := \prod_{\ell=1}^{p} U(H_B,\beta_\ell)U(H_C,\gamma_\ell) | s \rangle,\ \text{sendo}\
\mathbf{\gamma} = (\gamma_1, \dots, \gamma_p),\ \mathbf{\beta} = (\beta_1, \dots, \beta_p);
$$
iii) o valor esperado do hamiltoniano do custo com respeito ao estado evoluído:
$$
F(\mathbf{\gamma},\mathbf{\beta}) := \langle \mathbf{\gamma}, \mathbf{\beta} | H_C | \mathbf{\gamma}, \mathbf{\beta}  \rangle.
$$
O seguinte limite resolve o problema MaxCut:
$$
\underset{\mathbf{z} \in \left\{ 0, +1\right\}^{\times n_V}}{\arg \max} C(\mathbf{z})  = \lim_{p \rightarrow \infty} \underset{\mathbf{\gamma}, \mathbf{\beta}}{\arg \max }F(\mathbf{\gamma},\mathbf{\beta}).
$$
