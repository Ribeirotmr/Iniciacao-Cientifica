# Swap de emaranhamento

O Protocolo de Emaranhamento Swap mostra que é possível emaranhar dois sistemas quânticos, digamos A e B, que nunca interagiram diretamente. Isso pode acontece no protocolo porque os sistemas A e B estão inicialmente emaranhados com outros sistemas, digamos $C_1$ e $C_2$, os quais acabam sendo medidos (ou seja, interagindo). Portanto, o protocolo tem o efeito de trocar o emaranhamento de pares de sistemas quânticos mediante a interação de suas partes.


Circuito do protocolo:

![imagem_swap](https://drive.google.com/uc?export=view&id=1e2KvXP0c-al76uD6zIDMOzcsSUG08VLq)

Fonte:
[https://www.nature.com/articles/s41598-023-49326-4](https://www.nature.com/articles/s41598-023-49326-4)

**Swap de emaranhamento**: transferência dos emaranhamentos dos pares de qubits $q_0q_1$ e $q_2q_3$ para um emaranhamento do par $q_0q_3$.


**Estados de Bell**:

$$
\begin{split}
| \Phi_+\rangle = \frac{|00\rangle + | 11 \rangle}{\sqrt{2}}; \\
| \Phi_-\rangle = \frac{|00\rangle - | 11 \rangle}{\sqrt{2}}; \\
| \Psi_+\rangle = \frac{|01\rangle + | 10 \rangle}{\sqrt{2}}; \\
| \Phi_+\rangle = \frac{|01\rangle - | 10 \rangle}{\sqrt{2}}. \\
\end{split}
$$

**Cálculos**:
$$
\begin{split}
|00 \rangle_{q_0q_1} \otimes |00\rangle_{q_2q_3} \xrightarrow{H_0 H_2} &
	\frac{|0\rangle + |1\rangle}{\sqrt{2}}|0\rangle \otimes \frac{|0\rangle + |1\rangle}{\sqrt{2}}|0\rangle \\
    ... \xrightarrow{X_{0\rightarrow 1}^1 X_{2\rightarrow 1}^1} &  | \Phi_+\rangle_{q_0q_1} \otimes |\Phi_+\rangle_{q_2q_3} =
    \frac{|00\rangle_{q_0q_1} + |11\rangle_{q_0q_1}}{\sqrt{2}} \otimes \frac{|00\rangle_{q_2q_3} + |11\rangle_{q_2q_3}}{\sqrt{2}}\\
    &= \frac{1}{2}\left\lbrack |0000\rangle +|0011\rangle + |1100\rangle +|1111\rangle \right\rbrack_{q_0q_1q_2q_3}\\
    ... \xrightarrow{X_{1\rightarrow 2}^1} & \frac{1}{2}\left\lbrack |0000\rangle +|0011\rangle + |1110\rangle +|1101\rangle \right\rbrack_{q_0q_1q_2q_3}\\
    ... \xrightarrow{H_1}& \frac{1}{2\sqrt{2}}\left\lbrack |0\rangle (|0\rangle + |1\rangle) | 00\rangle +|0\rangle (|0\rangle + |1\rangle) |11\rangle + |1\rangle (|0\rangle - |1\rangle)|10\rangle +|1\rangle (|0\rangle - |1\rangle) 01\rangle \right\rbrack_{q_0q_1q_2q_3}\\
    &= \frac{1}{2\sqrt{2}}\left\lbrack |0000\rangle + |0100\rangle + |0011\rangle + |0111\rangle + |1010\rangle - |1110\rangle +|1001\rangle - |1101\rangle \right\rbrack_{q_0q_1q_2q_3}\\
    \underset{\color{green}{\text{(Troca das posições dos qubits: $q_0q_1q_2q_3 \rightarrow q_1q_2q_0q_1$)}}}{\leftrightharpoons} &= \frac{1}{2\sqrt{2}}\left\lbrack |0000\rangle + |1000\rangle + |0101\rangle + |1101\rangle + |0110\rangle - |1110\rangle +|0011\rangle - |1011\rangle \right\rbrack_{q_1q_2q_0q_3}\\
    &=\frac{1}{2\sqrt{2}}\left\lbrack |00\rangle (|00\rangle + |11\rangle) + |01\rangle (|01\rangle + |10\rangle ) + |10 \rangle (|00\rangle - |11\rangle) + |11\rangle(|01\rangle - |10\rangle) \right\rbrack_{q_1q_2q_0q_3}\\
    &=\frac{1}{2}\left\lbrack |00\rangle |\Phi_+\rangle + |01\rangle |\Psi_+\rangle + |10 \rangle |\Phi_-\rangle  + |11\rangle|\Psi_-\rangle) \right\rbrack_{q_1q_2q_0q_3}\\
    ... \xrightarrow[\color{green}{\text{(Medida dos qubits $q_1$ e $q_2$)}}]{\not{q_1}\not{q_2}} &
    \left\lbrace
    \begin{array}{l}
    |00\rangle_{q_1q_2} \frac{|00\rangle + | 11 \rangle_{q_0q_3}}{\sqrt{2}}\\
    |01\rangle_{q_1q_2} \frac{|01\rangle + | 10 \rangle_{q_0q_3}}{\sqrt{2}}\\
    |10 \rangle_{q_1q_2} \frac{|00\rangle - | 11 \rangle_{q_0q_3}}{\sqrt{2}}\\
    |11\rangle_{q_1q_2}\frac{|01\rangle - | 10 \rangle_{q_0q_3}}{\sqrt{2}}
    \end{array}
    \right.
    \ \ \ \ \ (1/4 = 25\% \ \text{de chance para cada alternativa})\\
    ... \xrightarrow{Z_{1\rightarrow 0}^1 X_{2\rightarrow 3}^1} &
    \left\lbrace
    \begin{array}{l}
    |00\rangle_{q_1q_2} \frac{|00\rangle + | 11 \rangle_{q_0q_3}}{\sqrt{2}}\\
    |01\rangle_{q_1q_2} \frac{|00\rangle + | 11 \rangle_{q_0q_3}}{\sqrt{2}}\\
    |10 \rangle_{q_1q_2} \frac{|00\rangle + | 11 \rangle_{q_0q_3}}{\sqrt{2}}\\
    |11\rangle_{q_1q_2} \frac{|00\rangle + | 11 \rangle_{q_0q_3}}{\sqrt{2}}
    \end{array}
    \right.
    \ \ \ \ = \ \ \ \
    \left\lbrace
    \begin{array}{l}
    |00\rangle_{q_1q_2} | \Phi_+\rangle_{q_0q_3}\\
    |01\rangle_{q_1q_2} | \Phi_+\rangle_{q_0q_3}\\
    |10\rangle_{q_1q_2} | \Phi_+\rangle_{q_0q_3}\\
    |11\rangle_{q_1q_2} | \Phi_+\rangle_{q_0q_3}
    \end{array}
    \right.\\
    &= \left\lbrack
        \frac{1}{4}|00\rangle\langle 00|_{q_1q_2} +
        \frac{1}{4}|01\rangle\langle 01|_{q_1q_2} +
        \frac{1}{4}|10\rangle\langle 10|_{q_1q_2} +
        \frac{1}{4}|11\rangle\langle 11|_{q_1q_2}
    \right\rbrack \otimes | \Phi_+\rangle\langle \Phi_+|_{q_0q_3}
\end{split}
$$

```python
# Qiskit: métodos básicos

from qiskit import QuantumCircuit, ClassicalRegister, QuantumRegister

from qiskit import transpile

from qiskit.visualization import plot_histogram, array_to_latex, plot_state_city
```

# Swap

```python
'''
Circuito Swap
'''

q = QuantumRegister(4,'q')
c = ClassicalRegister(4,'c')
qc = QuantumCircuit(q,c)
qc.h(q[0])
qc.h(q[2])
qc.cx(q[0],q[1])
qc.cx(q[2],q[3])
qc.barrier()
qc.cx(q[1],q[2])
qc.h(q[1])
qc.barrier()
# Medidas
qc.measure((q[1],q[2]),(c[1],c[2]))
qc.z(q[0]).c_if(c[1], 1)
qc.x(q[3]).c_if(c[2], 1)
qc.barrier()
qc.draw('mpl')
```

```python
'''
A medida dinâmica não é compatível com as funções "Operator" e "Statevector" do Qiskit.
Todavia, podemos verificar o estado do circuito até o ponto de as medidas serem realizadas.

Denominamos "circuito pré-Swap" a parte do circuito Swap desde o início até o ponto das medidas dinâmicas.
'''

'''
Circuito pré-Swap
'''

q = QuantumRegister(4,'q')
c = ClassicalRegister(4,'c')
qcp = QuantumCircuit(q,c)
qcp.h(q[0])
qcp.h(q[2])
qcp.cx(q[0],q[1])
qcp.cx(q[2],q[3])
qcp.barrier()
qcp.cx(q[1],q[2])
qcp.h(q[1])
qcp.barrier()
qcp.cx(q[0],q[3])
# Operador para definir a medida dos qubits q[0] e q[1] na base de Bell
qcp.h(q[0])
qcp.barrier()
display(qcp.draw('mpl'))

'''
Operador do circuito pré-Swap
'''

from qiskit.quantum_info import Operator
qcp_o = Operator(qcp)                      # NÃO FUNCIONA se houver medida dinâmica
print("Operador do circuito pré-Swap:")
display(qcp_o.draw('latex'))

'''
Estado gerado pelo circuito pré-Swap
'''

from qiskit.quantum_info import Statevector
state = Statevector.from_instruction(qcp) # NÃO FUNCIONA se houver medida dinâmica
print(state)
print("Estado:")
state.draw('latex')


```

```python
import numpy as np

# Definir uma tolerância numérica
tolerance = 1e-15

# Aplicar tolerância no operador do circuito (matriz unitaridade)
qcp_o_cleaned = np.where(np.abs(qcp_o.data) < tolerance, 0, qcp_o.data)

# Aplicar tolerância no estado vetorial
state_cleaned = np.where(np.abs(state.data) < tolerance, 0, state.data)

# Exibir os resultados limpos
print("Operador do circuito pré-Swap (com tolerância):")
print(qcp_o_cleaned)

print("\nEstado gerado pelo circuito pré-Swap (com tolerância):")
print(state_cleaned)
```

## Simulação Clássica

```python
'''
Simulação clássica do circuito Swap
'''

from qiskit_aer import AerSimulator

# Simulador
simulator = AerSimulator()

# Transpilação
qc_t = transpile(qc, simulator)

# Simulação
result = simulator.run(qc_t, shots = 2**10).result()

# Plotagem do histograma das estatísticas
counts = result.get_counts(qc_t)
plot_histogram(counts, title='Probabilidades na base computacional')
```
